```python
# Step 1: load the data 
import pandas as pd
df = pd.read_csv('Data/auto_mpg.csv') # note I am reading this data in from my data folder in the GitHub Repository
#df = pd.read_csv('/blue/zoo6927/share/Jupyter_Content/data/auto_mpg.csv') # Use this script if running on HyperGator Server
df # print data frame to make sure it looks okay
```

```python
# Step 2: Select the predictor variables for the model 

x = df[['cyl','disp','weight','accel']] #selected 6 variables
y = df[['mpg']]

print(x)
print(y)
```

```python
# Step 3 Create and fit the linear model
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(x,y)

print(model.intercept_, model.coef_) # print model coef
```

```python
# Step 4 print model evaluation information: R2 & MSE
import numpy as np

# First write function to return MSE
def mse(model, x, y):
    return np.mean((y - model.predict(x))**2)


print("MSE is", mse(model, x, y))
print("R2 is", model.score(x,y))
```
